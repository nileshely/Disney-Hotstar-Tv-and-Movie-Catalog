# Report on Disney+ Hotstar TV and Movie Catalog

Disney+ Hotstar, an Indian brand of subscription video on-demand streaming service, has emerged as a dominant force in the streaming landscape, offering a diverse catalog of films, television series, live sports, and original programming. Originally introduced as Hotstar, the platform gained immense popularity in India, capitalizing on the rapid growth of mobile broadband in the country. Following Disney's acquisition of 21st Century Fox in 2019, Hotstar was rebranded as Disney+ Hotstar in April 2020, integrating Disney's vast content library, including original programming and films from iconic brands such as Walt Disney Studios, Pixar, Marvel Studios, Lucasfilm, and National Geographic.

The data reveals key insights into the content available on Disney+ Hotstar, with over 6,000 titles categorized by genre, release year, age rating, running time, and type (TV show or movie). Visualizations showcase the distribution of age ratings, genres, and release years, providing a comprehensive overview of the platform's content diversity. Additionally, analyses of running time, seasons, and episodes shed light on viewing preferences and trends among subscribers. These insights enable Disney+ Hotstar to tailor its content strategy, catering to the preferences and interests of its audience effectively.

Furthermore, the report highlights the global expansion of Disney+ Hotstar beyond India, with operations in Indonesia, Malaysia, and Thailand, offering a blend of local and Disney+ content. The platform's presence in markets like Singapore, Canada, and the United Kingdom, targeting overseas Indians, underscores its strategic approach to catering to diverse audience segments. Moreover, the discontinuation of the U.S. version of Hotstar in 2021 reflects Disney's focus on consolidating its content offerings through domestic services like Hulu and ESPN+. This strategic shift aligns with evolving consumer preferences and market dynamics, ensuring Disney's continued competitiveness in the streaming industry.

In conclusion, Disney+ Hotstar's evolution from a regional streaming service to a global entertainment powerhouse underscores its adaptability and resilience in the ever-changing media landscape. With its extensive content library, strategic partnerships, and data-driven insights, Disney+ Hotstar continues to captivate audiences worldwide, offering a compelling and immersive streaming experience. As it continues to expand its footprint and innovate in content delivery, Disney+ Hotstar remains poised for sustained growth and success in the dynamic and competitive streaming market.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
hotstar_data = pd.read_csv("/kaggle/input/disney-hotstar-tv-and-movie-catalog/hotstar.csv")
```

```python
hotstar_data.head()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=hotstar_data, x='age_rating')
plt.title('Age Rating Distribution')
plt.xlabel('Age Rating')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.countplot(data=hotstar_data, y='genre', order=hotstar_data['genre'].value_counts().index)
plt.title('Genre Distribution')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=hotstar_data, x='type')
plt.title('Type Distribution (TV Show/Movie)')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=hotstar_data, x='year', bins=30)
plt.title('Release Year Distribution')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
top_titles = hotstar_data['title'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_titles.values, y=top_titles.index)
plt.title('Top 10 Titles by Count')
plt.xlabel('Count')
plt.ylabel('Title')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=hotstar_data, x='age_rating', y='running_time')
plt.title('Age Rating vs. Running Time')
plt.xlabel('Age Rating')
plt.ylabel('Running Time (minutes)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=hotstar_data, x='age_rating', hue='type')
plt.title('Age Rating vs. Type')
plt.xlabel('Age Rating')
plt.ylabel('Count')
plt.legend(title='Type')
plt.show()
```
